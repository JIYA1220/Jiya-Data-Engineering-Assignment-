# Section B: ETL Pipeline Implementation

## 1. Pipeline Architecture
The ETL pipeline is divided into three modular stages:
1. **Extraction**: Efficiently streaming raw CSV data into DuckDB.
2. **Transformation**: Cleaning, deduplicating, and deriving analytical columns.
3. **Loading**: Bulk loading data into a Star Schema with support for idempotency.


## 2. Transformation Logic

```
# Example of our Deduplication and Transformation Logic
# We use a single-pass CTE for maximum memory efficiency
'''
WITH filtered AS (
    SELECT * FROM raw_data
    WHERE event_time IS NOT NULL AND user_id IS NOT NULL
),
deduped AS (
    SELECT DISTINCT ON (user_id, product_id, event_time, event_type) * 
    FROM filtered
)
SELECT ... FROM deduped
'''
```

## 3. Incremental Loading & Idempotency

We ensure idempotency by using `INSERT OR IGNORE` and database-level sequences. This allows the pipeline to be re-run on the same files without creating duplicate records.
